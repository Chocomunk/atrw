```python
import os
import sagemaker

sagemaker_session = sagemaker.Session()

bucket = 'calvinandpogs-ee148'

role = sagemaker.get_execution_role()
print(bucket, role)
print(os.getcwd())
```

```python
from sagemaker.pytorch import PyTorch

estimator = PyTorch(entry_point='main.py',
                    source_dir='./',
                    framework_version='1.8.0',
                    role=role,
                    py_version='py3',
                    instance_count=1,
                    instance_type='ml.g4dn.xlarge',
                    hyperparameters={'dry-run': True})
```

```python
estimator.fit({'POSE': f's3://{bucket}/atrw/pose',
                'MODELS': f's3://{bucket}/models'})
#estimator.fit({'annot': f's3://{bucket}/atrw/pose/annotations/',
#               'train': f's3://{bucket}/atrw/pose/images/'})
```
